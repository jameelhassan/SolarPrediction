# Ensemble mean curve removing of data

```python
# Normalise the data
mu = np.mean(data)
sigma = np.std(data)
X = (data-mu)/sigma
```

```python
#Reshape data and plot of 1st day
datamat = X.reshape(366,-1)
print("The shape of the datamatrix is ", datamat.shape)
plt.plot(datamat[0])
plt.show()
```

```python
#Ensemble mean curve
ensmean = np.mean(datamat, axis=0)
ensmean.shape
plt.plot(ensmean)
```

```python
#Mean curve removed from each curve
datamat_norm = datamat-ensmean
plt.plot(datamat_norm[0])
```

```python
#Randomly shuffle rows to rearrange days
np.random.shuffle(datamat_norm)
plt.plot(datamat_norm[0])
```

```python
#Unroll randomly shuffled matrix of days
X = datamat_norm.reshape(-1,1)
```

```python
#Save the ensmean curve
savetxt("ensemble.csv", ensmean, delimiter=',')
```

**Using ensemble mean curve removed data as input**

```python
#Training, Testing data split
trainX_data = X[0:trainval]
testX_data = X[trainval:]
```

```python
#Plot of first training day and first testing day
print("These are the first training and testing day curves")
plt.plot(trainX_data[0:78])
plt.figure()
plt.plot(testX_data[0:78])
plt.show()
```

```python
#Function to Split into input ouput sets
def split_seq(seq, n_steps):
    X, y = list(), list()
    for i in range(len(seq)):
    
        #Find end of seq
        end_ix = i+n_steps
        #Check if sequence passed
        if end_ix>len(seq)-1:
            break

        #Gather input and ouputs
        seq_x, seq_y = seq[i:end_ix], seq[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
#No. of steps
n_steps = 4

#Split into inputs, outputs for training data
Xtrain, Ytrain = split_seq(trainX_data, n_steps)

n_features = 1
Xtrain = Xtrain.reshape(Xtrain.shape[0],Xtrain.shape[1], n_features)

#Split into input and outputs for test data
Xtest, Ytest = split_seq(testX_data, n_steps)

Xtest = Xtest.reshape(Xtest.shape[0], Xtest.shape[1], n_features)
```

```python
#Preparing Entire dataset for the network
Xfull, Yfull = split_seq(X, n_steps)
```

```python
#Dictionary holding test data for, test set and entire set for both normalised and difference t/f data
Yall = {'norm' : {'test' : Ytest, 'full' : Yfull}, 'diff' : {'test' : Ydiff_test, 'full' : Ydiff_full}}
Xall = {'norm' : {'test' : Xtest, 'full' : Xfull}, 'diff' : {'test' : Xdiff_test, 'full' : Xdiff_full}}
```

# Neural network model

```python
#Define Model
model = tf.keras.Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_steps,n_features)))
model.add(AveragePooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```


**Fit network model**

```python
#Fit model to mean removed data
hist = model.fit(Xtrain, Ytrain, epochs=50, batch_size=64, validation_split=0.15, verbose=0)
```

```python
#Fit model to difference transformed model
hist = model.fit(Xdiff_train, Ydiff_train, epochs=50, batch_size=64, validation_split=0.15, verbose=0)
```

**Predict using model**

```python
#Predicting normalised data using NN model
yhat_test = model.predict(Xall['norm']['test'], verbose=0)
yhat = model.predict(Xall['norm']['full'], verbose=0)
```

```python
#Predicting difference transformed data using NN model
yhat_difftest = model.predict(Xall['diff']['test'], verbose=0)
yhat_diff = model.predict(Xall['diff']['full'], verbose=0)
```
